# Setup

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score
```

# Loading Dataset

```python
data = pd.read_csv('FloodPrediction.csv')
data
```

```python
# don't need Sl column, case IDs are not helpful
data = data.drop('Sl', axis=1)
```

```python
coded_station_names = sorted(list(zip(coded_data['Station_Names'].unique(), data['Station_Names'].unique())))
coded_station_names
```

```python
coded_data.drop(['X_COR', 'Y_COR', 'Station_Number'], axis=1, inplace=True)
```

# Splitting and Engineering Data

```python
# split categorical and numeric features
x_category = coded_data.select_dtypes('category')
x_number = coded_data.select_dtypes('number').iloc[:, 0:-1]
y = coded_data.iloc[:, -1]

x_category, x_number, y
```

```python
# standardise numerical features
scaler = StandardScaler()
X_number = scaler.fit_transform(x_number)
X_number
```

```python
# concat categorical with scaled numerical features
X = np.concatenate((x_category.to_numpy(), X_number), axis=1)
X.shape
```

```python
# check feature importances (seed for reproducibility)
seed = 9352
forest = RandomForestClassifier(n_estimators=250,  random_state=seed)
forest.fit(X, y)
importances = forest.feature_importances_
sortedIndices = np.argsort(importances)
importances, sortedIndices
```

Based on feature importance analysis, we can see the order of importance in __descending order__ is:
 1) Rainfall
 2) Cloud_Coverage
 3) Bright_Sunshine
 4) Relative_Humidity
 5) Min_Temp
 6) Month
 7) Longitude
 8) Latitude
 9) Altitude
 10) Station_Name
 11) Max_Temp
 12) Wind Speed
 13) Period
 14) Year

Shows that the year and the period are not massively important, but the month is => the ___time___ of year is important _(seasonality)_

```python
# drop Year and Period
X = np.delete(X, [1, 13], 1)
X.shape
```

```python
# PCA of Cloud_Coverage and Bright_Sunshine
features = X[:, 7:8]
pca = PCA(n_components=1)
pca.fit(features)
pca_data = pca.transform(features)
combined_features = pca_data.flatten()
combined_features
```

```python
# retest importance with PCA result
X_pca = np.copy(X)
X_pca = np.delete(X_pca, np.s_[7:8], 1)
X_pca = np.concatenate((X_pca, combined_features.reshape(-1, 1)), axis=1)

forest = RandomForestClassifier(n_estimators=250,  random_state=seed)
forest.fit(X_pca, y)
importances = forest.feature_importances_
sortedIndices = np.argsort(importances)
importances, sortedIndices
```

Index 11 (the PCA feature) is still the second most important feature => PCA is good.

```python
X = np.copy(X_pca)
X.shape
```

```python
coded_data['Sunshine_Cloud_Coverage_PCA'] = combined_features
coded_data.drop(['Year', 'Period', 'Cloud_Coverage', 'Bright_Sunshine'], axis=1, inplace=True)

target = coded_data.pop('Flood')
coded_data.insert(len(coded_data.columns), 'Flood', target)
coded_data
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(coded_data.corr(), annot=True)
```

Correlations are significantly reduced, time to train and test model.

# Model Training

```python
train_features, test_features, train_labels, test_labels = model_selection.train_test_split(X, y, test_size=0.2, random_state=seed)
```

```python
# inital KNN train and test
knn = KNeighborsClassifier(n_neighbors=10, metric="euclidean")
knn = knn.fit(train_features, train_labels)
accuracy = knn.score(test_features, test_labels)
print("kNN Accuracy ", accuracy)
```

```python
# test multiple K values to determine best
knn_results = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k, metric="euclidean")
    knn.fit(train_features, train_labels)
    score = knn.score(test_features, test_labels)
    knn_results.append(score)

plt.xticks(range(1, 21, 1))
plt.xlabel("k Value")
plt.ylabel("Accuracy")
plt.plot(knn_results) 
```

Scores are all very good, but best result is at __k=4__. We will use this for cross-validation to ensure their is no overfitting.

```python
k_folds = 10

knn = KNeighborsClassifier(n_neighbors=4, metric="euclidean")
knn_scores = cross_val_score(knn, X, y, cv=k_folds)
knn_pred = cross_val_predict(knn, X, y, cv=k_folds)
knn_f_score = f1_score(y, knn_pred)
print(f"KNN Mean Accuracy: {knn_scores.mean()}, F Score: {knn_f_score}")
```

```python
x_range = np.arange(1, k_folds+1, 1)

plt.plot(x_range, knn_scores)
plt.title("K-Fold Cross Validation Score for KNN (k=5)")
plt.xlabel("k-Folds")
plt.xticks(x_range)
plt.ylabel("Accuracy")

plt.show()
```

Mean score stayed high, likely not overfitting and actually just a strong model. K-Fold results vary between 82% and 94%. 

F score after k-Folds is 0.715 which is average.
